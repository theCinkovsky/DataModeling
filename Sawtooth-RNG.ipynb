<center><font size = "5"> Simple sawtooth wave random number generator with quick compare to built-in solution </center>

```python
import numpy as np
from math import floor
import matplotlib.pylab as plt
import random
plt.rcParams["figure.figsize"]= 8, 8
random.seed(13)

# x0 != 0 !
x0 = 0.52; rng = []

# Care for periodicity of sawtooth wave based generator
teeth = 47.34
points = 875

def SawTooth(x_start, points):
    xn = x_start
    L = []
    for _ in range(points):
        xn1 = xn * teeth - floor(xn * teeth)
        L.append(xn1)
        xn = xn1        
    return L

Sawtooth_rng = SawTooth(x0, points)

np_rng = np.random.uniform(0, 1, points)

plt.subplot(221); plt.plot(np.linspace(0, 1, points), Sawtooth_rng, 'b*', markersize=3); plt.title('Sawtooth')
plt.subplot(222); plt.plot(np.linspace(0, 1, points), np_rng, 'r*', markersize=3); plt.title('Built-in')
plt.subplot(212); plt.hist(Sawtooth_rng, bins=100); plt.hist(np_rng, bins=100)

plt.show()
```
