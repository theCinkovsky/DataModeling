Random number generator based on inverse transform technique for:

* triangular PDF

\begin{equation}
f(x) = \left\{ \begin{array}{ll}
2x & \textrm{for x $\in$ [0, 1]}\\
0  & \textrm{for x $\in (-\infty, 0) \cup (1, \infty)$}\\
\end{array} \right.
\end{equation}

* exponential PDF

$$
f(x) = e^{-x}, x \geq 0
$$



```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams["figure.figsize"]= 8, 8
random.seed(13)

# for tests
points = 10000
xrand = np.random.uniform(0, 1, points)
xrand2 = np.random.uniform(0, 1, points)

xtria = lambda x: np.sqrt(x)
xexp = lambda x: -np.log(x)

plt.subplot(211); plt.hist(xtria(xrand), bins='auto'); plt.title('triangular ')
plt.subplot(212); plt.hist(xexp(xrand), bins='auto'); plt.title('exponential')
plt.show()
```

Laplace RNG using exponential RNG and additional variable $X \sim \mathcal{U}[0,1] $

\begin{equation}
\xi = \left\{ \begin{array}{ll}
1 & \textrm{for X $\geq$ 0.5}\\
-1 & \textrm{for X $< 0.5$}\\
\end{array} \right.
\end{equation}

```python
sign = lambda x: -1 * (x <= 0.5) + (x > 0.5)
lpc = lambda x, x2: -np.log(x) * sign(x2)

plt.hist(lpc(xrand, xrand2), bins='auto')
plt.show()
```
